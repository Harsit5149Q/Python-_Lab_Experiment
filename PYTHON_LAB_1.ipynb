```python
#12 Write a program to print truth table for bitwise operators (&, | and ^ operators)
print("Bitwise AND (&)")
print("A | B | A & B")
print("-" * 13)
print(f"0 | 0 | {0 & 0}")
print(f"0 | 1 | {0 & 1}")
print(f"1 | 0 | {1 & 0}")
print(f"1 | 1 | {1 & 1}")

print("\nBitwise OR (|)")
print("A | B | A | B")
print("-" * 13)
print(f"0 | 0 | {0 | 0}")
print(f"0 | 1 | {0 | 1}")
print(f"1 | 0 | {1 | 0}")
print(f"1 | 1 | {1 | 1}")

print("\nBitwise XOR (^)")
print("A | B | A ^ B")
print("-" * 13)
print(f"0 | 0 | {0 ^ 0}")
print(f"0 | 1 | {0 ^ 1}")
print(f"1 | 0 | {1 ^ 0}")
print(f"1 | 1 | {1 ^ 1}")
```

```python
#13 Write a program to find left shift and right shift values of a given number.
def bitwise_shifts(number, shift_count):
    left_shifted_value = number << shift_count
    right_shifted_value = number >> shift_count

    print(f"Original number: {number}")
    print(f"Shift count: {shift_count}")
    print("-" * 30)
    print(f"Left shift result ({number} << {shift_count}): {left_shifted_value}")
    print(f"Right shift result ({number} >> {shift_count}): {right_shifted_value}")
my_number = 20
my_shift = 2

bitwise_shifts(my_number, my_shift)
```

```python
#14 Using membership operator find whether a given number is in sequence (10,20,56,78,89)
sequence = (10, 20, 56, 78, 89)
number_to_find = 56
if number_to_find in sequence:
  print(f"{number_to_find} is in the sequence.")
else:
  print(f"{number_to_find} is not in the sequence.")
another_number = 99
if another_number in sequence:
  print(f"{another_number} is in the sequence.")
else:
  print(f"{another_number} is not in the sequence.")
```
