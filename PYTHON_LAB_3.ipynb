```python
#1 Find a factorial of given number.
number = int(input("Enter a number to calculate its factorial: "))
factorial = 1
if number < 0:
    print("Factorials do not exist for negative numbers.")
elif number == 0:
    print("The factorial of 0 is 1.")
else:
    for i in range(1, number + 1):
        factorial *= i
    print(f"The factorial of {number} is {factorial}")
```

```python
#2 Find whether the given number is Armstrong number.
num = int(input("Enter a number: "))
original_num = num
temp = num
n = 0
sum_of_powers = 0
while temp > 0:
    temp //= 10
    n += 1
temp = original_num
while temp > 0:
    digit = temp % 10
    power = 1
    for _ in range(n):
        power *= digit
    sum_of_powers += power
    temp //= 10
if sum_of_powers == original_num:
    print(f"{original_num} is an Armstrong number.")
else:
    print(f"{original_num} is not an Armstrong number.")
```

```python
#3 Print Fibonacci series up to given term.
n = int(input("Enter the number of terms: "))
a, b = 0, 1
if n <= 0:
    print("Please enter a positive integer")
elif n == 1:
    print(f"Fibonacci series: {a}")
else:
    print("Fibonacci series:", end=" ")
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
```

```python
#4 Write a program to find if given number is prime number or not.
num_input = input("Enter an integer to check: ")
num = int(num_input)
if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(f"{num} is not a prime number.")
            print(f"{i} is a factor of {num}.")
            break
    else:
        print(f"{num} is a prime number.")
elif num == 1:
    print(f"{num} is neither a prime nor a composite number.")
else:
    print(f"{num} is not a prime number.")

```

```python
#5 Check whether given number is palindrome or not.
num = int(input("Enter a number: "))
temp = num
reverse_num = 0
while temp > 0:
    digit = temp % 10
    reverse_num = (reverse_num * 10) + digit
    temp //= 10
if num == reverse_num:
    print(f"{num} is a palindrome number")
else:
    print(f"{num} is not a palindrome number")

```

```python
#7 Count and print all numbers divisible by 5 or 7 between 1 to 100.
count = 0
print("Numbers divisible by 5 or 7 (between 1 and 100):")
for num in range(1, 101):
    if (num % 5 == 0) or (num % 7 == 0):
        print(num, end=" ")
        count += 1
print(f"\nTotal count: {count}")
```

```python
#10 Write a program to print the following pattern
# 123454321
# 1234 *4321
# 123 * * 321
# 12 * * * 21
# 1 * * * * 1
rows = 5
for i in range(rows, 0, -1):
    for j in range(1, i + 1):
        print(j, end="")
    num_stars = rows - i
    if num_stars > 0:
        print(" " + "* " * (num_stars - 1) + "*", end="")
    start_desc = i
    if i == rows:
        start_desc = i - 1
    for j in range(start_desc, 0, -1):
        print(j, end="")
    print()
```

```python
#11 Write a program to print the sum of the following series
# 1+ ½ + 1/3 + ¼ +….+1/n
n = int(input("Enter the value of n: "))
total_sum = 0.0
for i in range(1, n + 1):
    total_sum += 1 / i
print(f"The sum of the series up to {n} terms is: {total_sum:.4f}")

```
