```python
#1  Write a Python function to find the maximum and minimum numbers from a sequence of numbers. (Note: Do not use built-in functions.)

def find_min_max(numbers):

    if not numbers:
        return None
    min_val = numbers[0]
    max_val = numbers[0]
    for num in numbers[1:]:
        if num < min_val:
            min_val = num
        elif num > max_val:
            max_val = num

    return (min_val, max_val)
my_list = [4, 7, 2, 9, 1, 5]
minimum, maximum = find_min_max(my_list)

print(f"The sequence is: {my_list}")
print(f"Minimum number: {minimum}")
print(f"Maximum number: {maximum}")
another_list = [2, 0, 2, 7, 5, -1, -2]
minimum_2, maximum_2 = find_min_max(another_list)

print(f"\nThe sequence is: {another_list}")
print(f"Minimum number: {minimum_2}")
print(f"Maximum number: {maximum_2}")
```

```python
#2 Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

def sum_of_cubes_smaller_than(n):

    if n <= 0:
        raise ValueError("Input must be a positive integer")
    sum_cubes = 0
    for i in range(1, n):
        sum_cubes += i**3
    return sum_cubes
number = 5
result = sum_of_cubes_smaller_than(number)
print(f"The sum of the cubes of integers smaller than {number} is: {result}")

number_2 = 10
result_2 = sum_of_cubes_smaller_than(number_2)
print(f"The sum of the cubes of integers smaller than {number_2} is: {result_2}")
```

```python
#3 Write a Python function to print 1 to n using recursion. (Note: Do not use loop)

def print_to_n_recursive(current_number, n):
    if current_number > n:
        return
    print(current_number)
    print_to_n_recursive(current_number + 1, n)
n_value = 10
print(f"Printing numbers from 1 to {n_value}:")
print_to_n_recursive(1, n_value)
```

```python
#4 Write a recursive function to print Fibonacci series upto n terms.
def recur_fibo(n):
    if n <= 1:
        return n
    else:
        return(recur_fibo(n-1) + recur_fibo(n-2))

nterms = 10
if nterms <= 0:
    print("Please enter a positive integer")
else:
    print("Fibonacci sequence:")
    for i in range(nterms):
        print(recur_fibo(i), end=" ")

```

```python
#5 Write a lambda function to find volume of cone

import math
volume_of_cone = lambda r, h: (1/3) * math.pi * r**2 * h
radius = 3.0
height = 5.0
volume = volume_of_cone(radius, height)
print(f"The volume of the cone with radius {radius} and height {height} is approximately {volume:.2f}")
```

```python
#6  Write a lambda function which gives tuple of max and min from a list.

get_max_min = lambda data_list: (max(data_list), min(data_list))
input_list = [10, 6, 8, 90, 12, 56]
output_tuple = get_max_min(input_list)
print(output_tuple)
```

```python
#8 Write a program to check whether all the values in a dictionary are the same or not using a lambda function.

dict1 = {'a': 1, 'b': 1, 'c': 1}
dict2 = {'x': 10, 'y': 20, 'z': 10}
empty_dict = {}
are_all_values_same = lambda d: len(set(d.values())) <= 1
print(f"Dictionary 1 values are all the same: {are_all_values_same(dict1)}")
print(f"Dictionary 2 values are all the same: {are_all_values_same(dict2)}")
print(f"Empty dictionary values are all the same: {are_all_values_same(empty_dict)}")
```

```python
#9 Write a program to create two lists and generate a dictionary with keys from list1 and values from list2.

list1 = ['name', 'age', 'city']
list2 = ['Alice', 30, 'New York']
dictionary_zip = dict(zip(list1, list2))

print("Dictionary using dict() and zip():")
print(dictionary_zip)
dictionary_comp = {key: value for key, value in zip(list1, list2)}

print("\nDictionary using dictionary comprehension:")
print(dictionary_comp)
dictionary_loop = {}
for i in range(len(list1)):
    dictionary_loop[list1[i]] = list2[i]

print("\nDictionary using a for loop:")
print(dictionary_loop)
```
